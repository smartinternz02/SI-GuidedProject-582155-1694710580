```python
#import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Univariate

```python
numerical_cols = ['total','speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']
```

```python
#histplot

for i in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[i], kde=True)
    plt.title(f'{i} Distribution')
    plt.show()
```

```python
#displot

for i in df.columns:
  sns.displot(df[i])
  plt.title(i)
  plt.show()
```

```python
#Histogram

for i in df.columns:
  plt.hist(df[i])
  plt.title(i)
  plt.show()
```

```python
# Boxplot

for i in numerical_cols:
  plt.boxplot(df[i])
  plt.title(i)
  plt.show()
```

### Bi variate

```python
#jointplot
for i in df.columns:
  sns.jointplot(x= df[i],y =df['abbrev'],data=df)
```

```python
#Line Chart
for i in df.columns:
  plt.plot(df[i],df['abbrev'])
  plt.xlabel(i)
  plt.ylabel("abbrev")
  plt.show()
```

```python
#Scatterplot
for i in df.columns:
  plt.scatter(df[i],df['abbrev'])
  plt.xlabel(i)
  plt.ylabel("abbrev")
  plt.show()
```

### Multi variate

```python
# Pairplot
sns.pairplot(df)
```
